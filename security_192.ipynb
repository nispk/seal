```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataframe_ckks = pd.read_csv(r'C:\Users\npkal\Desktop\SEAL_N\SEAL-master\bin\ckks\sec_level_192.csv')
```

```python
dataframe_bfv = pd.read_csv(r'C:\Users\npkal\Desktop\SEAL_N\SEAL-master\bin\bfv\sec_level_192.csv')
```

```python
labels = ['4096', '8192', '16384', '32768']
encode_time_ckks = dataframe_ckks.avg_encode_ckks
batchencode_time_bfv = dataframe_bfv.avg_batch_bfv
```

```python
x = np.arange(len(labels))
fig,ax = plt.subplots()
width = 0.35
rects1= ax.bar(x - width/2 , encode_time_ckks, width, label = 'ckks')
rects2= ax.bar(x + width/2 , batchencode_time_bfv, width, label = 'bfv')

ax.set_ylabel('Time(µs)')
ax.set_xlabel('Polynomial modulus degree')
ax.set_title('Encoding with security level 192 bits')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                   xytext =(0,3),
                   textcoords = "offset points",
                   ha = 'center', va = 'bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
fig.savefig('sec_192N\encode.png')
```
